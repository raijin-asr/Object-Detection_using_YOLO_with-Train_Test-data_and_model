```python
#Prediction on video
from ultralytics import YOLO
import cv2
import os  # Import the os module

# Get the directory where the code is executed
current_dir = os.getcwd()

model_path= current_dir + r"\object_results\train2\weights\best.pt"
model = YOLO(model_path)

# video detection
cap = cv2.VideoCapture("lambo2.mp4")

while True:
    ret, frame = cap.read()  
    
    if not ret:
        print("Error: Failed to open video")
        break

    # Predict objects on the frame
    predict_results = model.predict(frame, show=True)

    # Resize the frame to the desired width and height
    # frame = cv2.resize(frame, (1000, 600))
    # cv2.imshow("Object Detection in Video", frame)
    
    # Check for 'q' key press to close camera
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the frame and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
#Prediction on webcam

model_path= current_dir + r"\object_results\train2\weights\best.pt"
model = YOLO(model_path)

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Capture frame from webcam

    if not ret:
        print("Error: Failed to capture frame")
        break

    # Predict objects on the frame
    predict_results = model.predict(frame, show=True)

     # Resize the frame to the desired width and height
    frame = cv2.resize(frame, (1000, 600))
    cv2.imshow("Object Detection in Webcam", frame)
    
    # Check for 'q' key press to close camera
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
